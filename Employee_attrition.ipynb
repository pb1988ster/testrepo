```python
df_small = df[['Attrition','Age', 'DistanceFromHome', 'HourlyRate', 'PercentSalaryHike', 'NumCompaniesWorked','PerformanceRating','StockOptionLevel', 'TotalWorkingYears', 'JobSatisfaction', 'MonthlyIncome', 'YearsAtCompany', ]]

df_small['Attrition'] = df_small['Attrition'].apply(lambda x: 1 if x=='Yes' else 0)

df_small.head()
```

```python
y = df_small['Attrition']
y
```

```python
# sklearn - machine learning library.

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Decision Trees (Black Box)
from sklearn.tree import DecisionTreeClassifier # ML Algorithm or Model.
```

```python
model = DecisionTreeClassifier(max_depth = 4) # creating a model
model.fit(X_train, y_train) # training of the ML Model
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred[:10]
```

```python
# acutal
y_test[:10]
```

```python
# 85% is the accuracy
np.mean(y_pred == y_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
